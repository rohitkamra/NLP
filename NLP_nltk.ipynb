#### Tokenization- Sentences and Word wise

---




````
sentences = nltk.sent_tokenize(trump_txt)
sentences
````

````
words = nltk.word_tokenize(trump_txt)
words
````

#### **Stemming**: Find the base word or get the root word from verbs, some of the words dont give meaningful word, Apps- Spam Classifier, Sentiment classifier
***** Create Intermidate representation

---


#### **Lemmatization**: Find the meaningful word, work is same like Stemming, takes more time compare to stemming Apps- Chatbot, QnA apps where human intervention



````
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
````

````
stemmer = PorterStemmer()
````

````
stopwords.words('english')
````

````
newSentences = []
for sentIndex in range(len(sentences)):
  words = nltk.word_tokenize(sentences[sentIndex])
  words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
  newSentences.append(" ".join(words))

for i in range(len(sentences)):
  print('Old Sentence : ', sentences[i])
  print('New Sentence : ', newSentences[i])
````

````
from nltk.stem import WordNetLemmatizer
````

````
lemm = WordNetLemmatizer()
````

````
newSentencesLemm = []
for sentIndex in range(len(sentences)):
  words2 = nltk.word_tokenize(sentences[sentIndex])
  words2 = [lemm.lemmatize(word) for word in words2 if word not in set(stopwords.words('english'))]
  newSentencesLemm.append(" ".join(words2))

for i in range(len(sentences)):
  print('Old Sentence : ', sentences[i])
  print('New Sentence : ', newSentencesLemm[i])
````

#### Bag of words: 
1. Always lower the words
2. stemming,  lemmetization, stop keywords
3. Count frequency of words and put in decending order sorting
4. Give value 1 if word exist otherwise 0, sentence wise

Disadvantage- Give equal weightage to each word
To solve this Use TFIDF and for Huge Dataset too 

---



````
import re
````

````
corpus = []

for sentIndex in range(len(sentences)):
  sentence = re.sub(r"[^a-zA-Z]", " ", sentences[sentIndex])
  sentence = sentence.lower().split()
  words = [lemm.lemmatize(word) for word in sentence if word not in set(stopwords.words('english'))]
  sentence  = " ".join(words)
  corpus.append(sentence)
corpus
````

Create Bag of words using scikit-learn

````
from sklearn.feature_extraction.text import CountVectorizer
````

````
cv = CountVectorizer()
X = cv.fit_transform(corpus).toarray()
X
````

#### TF-IDF: Term Frequency and Inverse Document Frequency

```
TF = no of repetitive words in sentence / no of words in sentence 

IDF =  log(No of sentences / No of sentences containing words)

Final - TF * IDF
```



````
from sklearn.feature_extraction.text import TfidfVectorizer
tv = TfidfVectorizer()
X = tv.fit_transform(corpus).toarray()
X
````
